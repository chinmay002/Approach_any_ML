```python
import random
import datetime


cat1 = [random.randint(1, 5) for _ in range(10)]
cat2 = [random.randint(1, 3) for _ in range(10)]
cat3 = [random.randint(0,2) for _ in range(10)]
num1 = [random.uniform(-2,2) for _ in range(10)]


start_date = datetime.date(2021, 1, 1)
end_date = datetime.date(2023, 12, 31)

month_range = range(1, 13)

year_range = range(start_date.year, end_date.year+1)

random_dates = [datetime.date(year=random.choice(year_range), month=random.choice(month_range), day=1) + datetime.timedelta(days=random.randint(0, 365)) for _ in range(10)]

num_digits = 5

customer_id = [random.randint(11, 20) for _ in range(10)]

df = pd.DataFrame({'date':random_dates,'customer_id':customer_id,'cat1':cat1,'cat2':cat2,'cat3':cat3,'num1':num1})
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df1 = df.copy()
```

```python
df1['month'] = df['date'].dt.month
df1['year'] = df['date'].dt.year
df1['weekofyear'] = df['date'].dt.weekofyear
df1['dayofweek'] = df['date'].dt.dayofweek
df1['weekend'] = (df['date'].dt.weekday>=5).astype(int)
```
