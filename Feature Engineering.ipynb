```python
def generate_features(df):
    df.loc[:,'month'] = df['date'].dt.month
    df.loc[:,'year'] = df['date'].dt.year
    df.loc[:,'weekofyear'] = df['date'].dt.weekofyear
    df.loc[:,'dayofweek'] = df['date'].dt.dayofweek
    df.loc[:,'weekend'] = (df['date'].dt.weekday>=5).astype(int)
    
    aggs ={}
    aggs['month'] = ['nunique','mean']        #df.agg({'col':['sum','mean']})
    aggs['weekofyear'] = ['nunique','mean']
    aggs['num1']= ['sum','mean','max','min']
    aggs['customer_id'] = ['size']            #get count of customer_id 
    
    agg_df = df.groupby('customer_id').agg(aggs)  #cust_idx becomes index col
    return agg_df.rest_index()               #to make customer_idx non index col
```

# Feature Selection

If the features have a very low variance (i.e. very close to 0), they
are close to being constant and thus, do not add any value to any model at all. It
would just be nice to get rid of them and hence lower the complexity

Univariate selection.
<li>scoreing each vol against target col.such as Mutual Info, ANOVA, F-test,chi2</li>
    <li>using SelectKBest
    <li>Chi2 used only when non-negative in nature.

```python
class UnivariateFeatureSelection():
    def __init__(self,n_feature,problem_type,scoring):
        '''
        custom wrapper on different univariate features selction models of sklearn
        :param n_feature: SelectPercentile if float else SelecKBest
        :param problem_type : classification or regression
        :param scoring :scoring_function
        
        '''
        
        if problem_type == 'classification':
            valid_scoring = {
                'f_classif':f_classif,
                'chi2':chi2,
                'mutual_info_classif':mutual_info_classif
            }
        else:
            valid_scoring = {
                'f_regression':f_regression,
                'mutual_info_regression':mutual_info_regression
            }
        if scoring not in valid_scoring:
            raise Exception('Invalid Scoring Function')
        
        if isinstance(n_features,int):
            self.selection = SelectKBest(valid_scoring[scoring],k=n_features)
        elif isinstance(n_features,float):
            self.selection = SelectPercentile(valid_scoring[scoring],
                                             percentile=int(n_features*100))
        else:
            raise Exception('Invalid type of feature')
            
    def fit(self,X,y):
        return self.selection.fit(X,y)
 
        # same transform function
    def transform(self, X):
        return self.selection.transform(X)
    
    # same fit_transform function
    def fit_transform(self, X, y):
        return self.selection.fit_transform(X, y)
```
